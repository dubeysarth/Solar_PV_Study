```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
url = lambda filename: os.path.join(os.getcwd(), 'raw_data', f"{filename}.csv")
```

```python
Sampling_Cycles = pd.read_csv(url('Sampling_Cycles'))
Sampling_Cycles['Date'] = pd.to_datetime(Sampling_Cycles['Date'])
print("Done: Sampling_Cycles")

Meteorological_Data = pd.read_csv(url('Meteorological_Data'))
Meteorological_Data.dropna(axis = 0, how = 'all', inplace=True)
Meteorological_Data['DateTime'] = pd.to_datetime(Meteorological_Data['Date'] + ' ' + Meteorological_Data['Time'])
# # Meteorological_Data = Meteorological_Data.drop(['Date', 'Time'], axis=1)
print("Done: Meteorological_Data")

PM_Data = pd.read_csv(url('PM_Data'))
PM_Data.dropna(axis = 0, how = 'all', inplace=True)
PM_Data.columns = [x.strip() for x in PM_Data.columns]
PM_Data['DateTime'] = pd.to_datetime(PM_Data['Date'] + ' ' + PM_Data['Time'])
# PM_Data = PM_Data.drop(['Date', 'Time'], axis=1)
print("Done: PM_Data")

LAMPS_FT_Raw_Data = pd.read_csv(url('LAMPS_FT_Raw_Data'))
LAMPS_FT_Raw_Data.dropna(axis = 0, how = 'all', inplace = True)
LAMPS_FT_Raw_Data.columns = [x.strip() for x in LAMPS_FT_Raw_Data.columns]
LAMPS_FT_Raw_Data['DateTime'] = pd.to_datetime(LAMPS_FT_Raw_Data['Date'] + ' ' + LAMPS_FT_Raw_Data['Time'])
# LAMPS_FT_Raw_Data = LAMPS_FT_Raw_Data.drop(['Date', 'Time'], axis=1)
print("Done: LAMPS_FT_Raw_Data")

LAMPS_DN_Raw_Data = pd.read_csv(url('LAMPS_DN_Raw_Data'))
LAMPS_DN_Raw_Data.dropna(axis = 0, how = 'all', inplace = True)
LAMPS_DN_Raw_Data.columns = [x.strip() for x in LAMPS_DN_Raw_Data.columns]
LAMPS_DN_Raw_Data['DateTime'] = pd.to_datetime(LAMPS_DN_Raw_Data['Date'] + ' ' + LAMPS_DN_Raw_Data['Time'])
# LAMPS_DN_Raw_Data = LAMPS_DN_Raw_Data.drop(['Date', 'Time'], axis=1)
print("Done: LAMPS_DN_Raw_Data")

CampSci_Reference_Panel = pd.read_csv(url('CampSci_Reference_Panel'))
CampSci_Reference_Panel.dropna(axis = 0, how = 'all', inplace = True)
CampSci_Reference_Panel.columns = [x.strip() for x in CampSci_Reference_Panel.columns]
CampSci_Reference_Panel['TIMESTAMP'] = pd.to_datetime(CampSci_Reference_Panel['TIMESTAMP'])
print("Done: CampSci_Reference_Panel")

Day_Fraction = pd.read_csv(url('Day_Fraction'))
Day_Fraction.dropna(axis = 0, how = 'all', inplace = True)
Day_Fraction.columns = [x.strip() for x in Day_Fraction.columns]
Day_Fraction['Date'] = pd.to_datetime(Day_Fraction['Date'])
print("Done: Day_Fraction")

LAMPS_Calculated_Soiling_Loss = pd.read_csv(url('LAMPS_Calculated_Soiling_Loss'))
LAMPS_Calculated_Soiling_Loss.dropna(axis = 0, how = 'all', inplace = True)
LAMPS_Calculated_Soiling_Loss.columns = [x.strip() for x in LAMPS_Calculated_Soiling_Loss.columns]
LAMPS_Calculated_Soiling_Loss['Date'] = pd.to_datetime(LAMPS_Calculated_Soiling_Loss['Date'])
print("Done: LAMPS_Calculated_Soiling_Loss")

Additional_Data = pd.read_csv(url('Additional_Data'))
Additional_Data.dropna(axis = 0, how = 'all', inplace = True)
Additional_Data.columns = [x.strip() for x in Additional_Data.columns]
Additional_Data['Date'] = pd.to_datetime(Additional_Data['Date'])
print("Done: Additional_Data")
```

# Meteorological Data

```python
Meteorological_Data = pd.read_csv(url('Meteorological_Data'))
Meteorological_Data.dropna(axis = 0, how = 'all', inplace=True)
Meteorological_Data['DateTime'] = pd.to_datetime(Meteorological_Data['Date'] + ' ' + Meteorological_Data['Time'])
Meteorological_Data = Meteorological_Data.drop(['Date', 'Time', 'RECORD'], axis=1)
Meteorological_Data.set_index('DateTime', inplace=True)
Meteorological_Data = Meteorological_Data.resample('1T').interpolate(method='linear')
Meteorological_Data['Date'] = Meteorological_Data.index.date
Meteorological_Data['Time'] = Meteorological_Data.index.time
Meteorological_Data = Meteorological_Data[(Meteorological_Data['Date'] >= daily_dates[0]) & (Meteorological_Data['Date'] <= daily_dates[-1])]
print(Meteorological_Data.shape, Meteorological_Data.index[0], Meteorological_Data.index[-1])
Meteorological_Data
```

```python
# Meteorological_Data
# Reset index and drop DateTIme
Meteorological_Data.reset_index(inplace=True)
Meteorological_Data.drop(['DateTime'], axis=1, inplace=True)
# In Time column, set seconds to 0
Meteorological_Data['Time'] = Meteorological_Data['Time'].apply(lambda x: x.replace(second=0))
# Set index to Date and Time
Meteorological_Data.set_index(['Date', 'Time'], inplace=True)
Meteorological_Data
```

```python
PM_Data = pd.read_csv(url('PM_Data'))
PM_Data.dropna(axis = 0, how = 'all', inplace=True)
PM_Data.columns = [x.strip() for x in PM_Data.columns]
PM_Data['DateTime'] = pd.to_datetime(PM_Data['Date'] + ' ' + PM_Data['Time'])
PM_Data = PM_Data.drop(['Date', 'Time'], axis=1)
PM_Data.set_index('DateTime', inplace=True)
PM_Data['Date'] = PM_Data.index.date
PM_Data['Time'] = PM_Data.index.time
PM_Data = PM_Data[(PM_Data['Date'] >= daily_dates[0]) & (PM_Data['Date'] <= daily_dates[-1])]
print(PM_Data.shape, PM_Data.index[0], PM_Data.index[-1])
PM_Data
```

```python
# PM_Data
# Reset index and drop DateTIme
PM_Data.reset_index(inplace=True)
PM_Data.drop(['DateTime'], axis=1, inplace=True)
# In Time column, set seconds to 0
PM_Data['Time'] = PM_Data['Time'].apply(lambda x: x.replace(second=0))
# Set index to Date and Time
PM_Data.set_index(['Date', 'Time'], inplace=True)
PM_Data
```

```python
# LAMPS_FT_Raw_Data
# Reset index and drop DateTIme
LAMPS_FT_Raw_Data.reset_index(inplace=True)
LAMPS_FT_Raw_Data.drop(['DateTime'], axis=1, inplace=True)
# In Time column, set seconds to 0
LAMPS_FT_Raw_Data['Time'] = LAMPS_FT_Raw_Data['Time'].apply(lambda x: x.replace(second=0))
# Set index to Date and Time
LAMPS_FT_Raw_Data.set_index(['Date', 'Time'], inplace=True)
LAMPS_FT_Raw_Data
```

```python
CampSci_Reference_Panel = pd.read_csv(url('CampSci_Reference_Panel'))
CampSci_Reference_Panel.dropna(axis = 0, how = 'all', inplace = True)
CampSci_Reference_Panel.columns = [x.strip() for x in CampSci_Reference_Panel.columns]
CampSci_Reference_Panel['TIMESTAMP'] = pd.to_datetime(CampSci_Reference_Panel['TIMESTAMP'])
CampSci_Reference_Panel['GeffRef'] = CampSci_Reference_Panel['GeffRef'].astype(float)
CampSci_Reference_Panel['TempRef'] = CampSci_Reference_Panel['TempRef'].astype(float)
CampSci_Reference_Panel = CampSci_Reference_Panel[(CampSci_Reference_Panel['TIMESTAMP'] >= daily_dates[0]) & (CampSci_Reference_Panel['TIMESTAMP'] <= daily_dates[-1])]
CampSci_Reference_Panel
```

```python
CampSci_Reference_Panel.info()
```

```python
# Take mean of repeating TIMESTAMPs and make a new dataframe
CampSci_Reference_Panel_Mean = CampSci_Reference_Panel.groupby('TIMESTAMP').mean()
CampSci_Reference_Panel_Mean
```

```python
# CampSci_Reference_Panel_Mean set 2 index columns Date and Time, and drop TIMESTAMP
CampSci_Reference_Panel_Mean['Date'] = CampSci_Reference_Panel_Mean.index.date
CampSci_Reference_Panel_Mean['Time'] = CampSci_Reference_Panel_Mean.index.time
# reset index
CampSci_Reference_Panel_Mean.reset_index(inplace=True)
CampSci_Reference_Panel_Mean = CampSci_Reference_Panel_Mean.drop(['TIMESTAMP'], axis=1)
# Set index to both Date and Time
CampSci_Reference_Panel_Mean.set_index(['Date', 'Time'], inplace=True)
CampSci_Reference_Panel_Mean
```

```python
CampSci_Reference_Panel_Mean = CampSci_Reference_Panel_Mean.groupby(['Date', 'Time']).mean()
CampSci_Reference_Panel_Mean
```

```python
# Combine LAMPS_FT_Raw_Data and LAMPS_DN_Raw_Data
Combined_Data = pd.concat([LAMPS_FT_Raw_Data, LAMPS_DN_Raw_Data], axis=1)
Combined_Data
```

```python
# Combined_Date take mean of repeating Date and Time
Combined_Data_Mean = Combined_Data.groupby(['Date', 'Time']).mean()
Combined_Data_Mean
```

```python
# Take mean of repeating Date and Time
PM_Data_Mean = PM_Data.groupby(['Date', 'Time']).mean()
Meteorological_Data_Mean = Meteorological_Data.groupby(['Date', 'Time']).mean()
print(PM_Data_Mean.shape, Meteorological_Data_Mean.shape)
```

```python
# Combine Combined_Data and PM_Data and Meteorological_Data
Combined_Data = pd.concat([Combined_Data_Mean, PM_Data_Mean, Meteorological_Data_Mean], axis=1)
Combined_Data
```

```python
# Combine Combined_Data and CampSci_Reference_Panel_Mean
Combined_Data = pd.concat([Combined_Data, CampSci_Reference_Panel_Mean], axis=1)
Combined_Data
```
