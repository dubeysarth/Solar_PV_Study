```python
import numpy as  np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import os
import glob
import itertools
from tqdm import tqdm
```

```python
Master = []
```

```python
# Get list of sheets within excel file
def get_sheet_names(file_path):
    xl = pd.ExcelFile(file_path)
    return xl.sheet_names

get_sheet_names(r'Final Master Data Updated.xlsx')
```

```python
df = pd.read_excel(r'Final Master Data Updated.xlsx', sheet_name='LAMPS DN Raw Data', skiprows=0, dtype = str)
# Remove rows with all NaN values
df = df.dropna(how='all')
df_dates = df.iloc[:, 1:-3]
df_dates.columns = np.arange(0, len(df_dates.columns))
df_dates
```

```python
col_idx = 11
```

```python
df1 = df_dates.iloc[:, col_idx].values
df1
```

```python
df2 = [x.split('/') if not ' ' in x else x.split(' ')[0].split('-') for x in df1 if type(x) == str]
df2
```

```python
np.unique(np.array(df2), axis=0, return_counts=True)
```

```python
# Rearrange numpy columns 
df3 = np.array(df2)
df3[:, [0, 1, 2]] = df3[:, [1, 0, 2]] # Day, Month, Year
df3
```

```python
# df3[:, 1] = '4'
# df3[:, 2] = '2022'
# df3
```

```python
# Merge along row and turn to datetime
df4 = pd.to_datetime(pd.DataFrame(df3).apply(lambda x: '-'.join(x), axis=1), format='%d-%m-%Y')
df4
```

```python
Master += list(df4.values)
```
